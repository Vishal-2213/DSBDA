```python
# 1. Create data subsets for type of post

photo = data[data["Type"] == 'Photo']
status = data[data["Type"] == 'Status']
link = data[data["Type"] == 'Link']
video = data[data["Type"] == 'Video']
```

```python
# 2. Merge two subsets

merged = pd.concat([photo, status])
```

```python
# 3. Sort data on total page likes

data.sort_values(by="Page total likes")
```

```python
# 4. Transposing dataframe

data.transpose()
```

```python
# 5. Melting dataset into long format

melt_data = data[["Page total likes", "Type", "Category"]]
```

```python
melt_data
```

```python
pd.melt(melt_data, id_vars=["Type", "Category"])
```

```python
# 6. Casting data into wide format

new_data = melt_data.drop_duplicates()
new_data
```

```python
data_new = new_data.sort_values(by="Page total likes")
```
